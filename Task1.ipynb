```python
# storing correct username and password
username = "admin"
password = "1234"

# taking input from user
enter_username = input("Enter username: ")
enter_password = input("Enter password: ")

# checking whether both username and password are correct
if enter_username == username and enter_password == password:
    print("Login Successful")   # login success message
else:
    print("Invalid Credentials")   # if details are wrong
```

```python
# list of student marks
marks = [45, 78, 90, 33, 60]

# counters for pass and fail
pass_count = 0
fail_count = 0

# checking each mark
for m in marks:
    if m >= 50:     # if marks greater than or equal to 50
        pass_count += 1
    else:
        fail_count += 1

# printing final result
print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

```python
# list of names with extra spaces
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

# cleaning each name
for name in names:
    cleaned = name.strip().lower()   # removing spaces and converting to lowercase
    cleaned_names.append(cleaned)    # adding cleaned name to new list

print("Cleaned Names:", cleaned_names)
```

```python
# list of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# checking each message
for msg in messages:
    length = len(msg)   # finding length of message
    print("Message:", msg)
    print("Length:", length)

    # checking if message length is more than 10
    if length > 10:
        print("This message is longer than 10 characters")

    print()   # for spacing
```

```python
# list of system logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0  # counter for errors

# checking each log
for log in logs:
    if log == "ERROR":   # if log shows error
        error_count += 1

# printing total error count
print("Total ERROR messages:", error_count)
```
