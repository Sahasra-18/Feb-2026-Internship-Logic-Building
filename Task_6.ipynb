```python
# Smart Parking Lot Management System

# parking capacity
capacity = 50

# vehicle entry and exit logs
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

# variable to store current parked vehicles
parked = 0

# variable to store peak parking usage
peak = 0

# loop through each log
for log in vehicle_logs:

    # if vehicle enters
    if log == "IN":
        parked = parked + 1

    # if vehicle exits
    elif log == "OUT":
        if parked > 0:      # avoid negative count
            parked = parked - 1

    # check peak usage
    if parked > peak:
        peak = parked

# check parking status
if parked > capacity:
    status = "ALERT: Parking Full"
else:
    status = "Available"

# display results
print("Currently Parked Vehicles:", parked)
print("Peak Parking Usage:", peak)
print("Parking Status:", status)
```

```python
# Online Food Delivery Time Estimator

# function to calculate delivery time
def delivery_time(distance, traffic, weather):

    # base time calculation
    # assume 1 km takes 5 minutes normally
    time = distance * 5

    # traffic condition
    if traffic == "High":
        time = time + 10      # add 10 minutes delay
    elif traffic == "Medium":
        time = time + 5       # add 5 minutes delay
    else:
        time = time           # no delay

    # weather condition
    if weather == "Rainy":
        time = time + 5       # extra delay due to rain
    elif weather == "Stormy":
        time = time + 15      # heavy delay

    # print final estimated time
    print("Estimated Delivery Time:", time, "minutes")


# Sample Input
distance = 8
traffic = "High"
weather = "Rainy"

# function call
delivery_time(distance, traffic, weather)
```

```python
# Movie Theatre Seat Occupancy Analyzer

# function to analyze seat occupancy
def seat_analyzer(total_seats, booked_seats):

    # count how many seats are booked
    booked_count = len(booked_seats)

    # calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    # check show status
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 70:
        status = "Almost Full"
    else:
        status = "Seats Available"

    # suggestion for additional shows
    if occupancy >= 90:
        suggestion = "Consider Opening Additional Show"
    else:
        suggestion = "No Extra Show Needed"

    # display results
    print("Occupancy:", int(occupancy), "%")
    print("Show Status:", status)
    print("Suggestion:", suggestion)


# Sample Input
total_seats = 200

# creating 150 booked seats (sample data)
booked_seats = [1] * 150

# function call
seat_analyzer(total_seats, booked_seats)
```

```python
# Cloud Server Load Classification System

# function to classify server load
def server_load(cpu_readings):

    total = 0   # variable to store sum of readings

    # loop to add all CPU readings
    for value in cpu_readings:
        total = total + value

    # calculate average CPU usage
    average = total / len(cpu_readings)

    # classify server status
    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else:
        status = "Critical"

    # display result
    print("Average CPU Load:", int(average), "%")
    print("Server Status:", status)


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

# function call
server_load(cpu_readings)
```

```python
# Smart Classroom Resource Usage Monitor

# function to check resource usage
def resource_monitor(resources):

    overused = []   # list to store overused resources

    # assume usage above 8 hours is overuse
    limit = 8

    # check each resource using loop
    for item in resources:
        if resources[item] > limit:
            overused.append(item)

    # display overused resources
    if len(overused) > 0:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

# function call
resource_monitor(resource_usage)
```

```python
# Online Event Registration Capacity Controller

# function to manage registrations
def event_registration(capacity, registrations):

    confirmed = 0   # confirmed participants
    waitlist = 0    # waitlisted users

    # loop through each registration
    for i in range(registrations):

        # check if seats are available
        if confirmed < capacity:
            confirmed = confirmed + 1
        else:
            waitlist = waitlist + 1

    # registration status
    if confirmed == capacity:
        status = "Closed"
    else:
        status = "Open"

    # display results
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)


# Sample Input
capacity = 100
registrations = 105

# function call
event_registration(capacity, registrations)
```
