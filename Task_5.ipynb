```python
# Function to check post engagement
def post_engagement(likes_list):

    total_likes = 0   # variable to store total likes

    # loop to add all likes
    for like in likes_list:
        total_likes = total_likes + like

    # print total likes
    print("Total Likes:", total_likes)

    # check if post is viral or normal
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# list of likes on different posts
likes = [200, 150, 300, 250, 150]

# calling the function
post_engagement(likes)
```

```python
# Function to check rainfall adequacy
def rainfall_checker(rainfall_list, required_level):

    total_rainfall = 0   # variable to store total rainfall

    # loop to add rainfall values
    for rain in rainfall_list:
        total_rainfall = total_rainfall + rain

    # calculate average rainfall
    average_rainfall = total_rainfall / len(rainfall_list)

    # print average rainfall
    print("Average Rainfall:", int(average_rainfall))

    # check rainfall status
    if average_rainfall >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


# example seasonal rainfall data
rainfall_data = [70, 75, 68, 80, 67]

# required rainfall level
required = 70

# calling the function
rainfall_checker(rainfall_data, required)
```

```python
# Function to check duplicate usernames
def check_duplicates(usernames):

    # convert list into set (set removes duplicates)
    unique_users = set(usernames)

    # compare lengths to find duplicates
    if len(unique_users) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# list of usernames
users = ["user1", "user2", "user3", "user1", "user4"]

# calling the function
check_duplicates(users)
```

```python
# Function to check appointment eligibility
def check_eligibility(age):

    # print patient age
    print("Patient Age:", age)

    # check eligibility condition
    if age >= 18:
        print("Eligibility Status: Eligible")
    else:
        print("Eligibility Status: Not Eligible")


# example patient age
patient_age = 21

# calling the function
check_eligibility(patient_age)
```

```python
# Function to filter premium crop prices
def premium_crop_filter(prices):

    premium_crops = []   # list to store premium prices

    # loop through all crop prices
    for price in prices:
        # check if price is greater than 2000
        if price > 2000:
            premium_crops.append(price)   # add to premium list

    # display premium crops
    print("Premium Crops:", premium_crops)


# list of crop prices
crop_prices = [1500, 2500, 1800, 3200, 1900]

# calling the function
premium_crop_filter(crop_prices)
```

```python
# Function to check application health
def check_health(errors):

    # print error count
    print("Error Count:", errors)

    # check system status using conditions
    if errors == 0:
        print("System Status: Healthy")
    elif errors <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


# example error count
error_count = 7

# calling the function
check_health(error_count)
```

```python
# Function to check daily transaction limit
def check_transaction(amount):

    daily_limit = 50000   # daily transaction limit

    # print transaction amount
    print("Transaction Amount:", amount)

    # check if amount is within limit
    if amount <= daily_limit:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")


# example transaction amount
transaction_amount = 60000

# calling the function
check_transaction(transaction_amount)
```

```python
# Function to check attendance eligibility
def check_attendance(attendance_list):

    total_classes = len(attendance_list)   # total number of classes
    present_count = 0   # variable to count present days

    # loop to count attendance
    for day in attendance_list:
        if day == 1:        # 1 means present
            present_count = present_count + 1

    # calculate attendance percentage
    percentage = (present_count / total_classes) * 100

    # print attendance percentage
    print("Attendance Percentage:", percentage)

    # check eligibility
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# attendance data (1 = present, 0 = absent)
attendance = [1, 1, 0, 1, 1]

# calling the function
check_attendance(attendance)
```

```python
# Function to calculate electricity bill
def electricity_bill(units):

    bill = 0   # variable to store total bill

    # calculate bill using slab system
    if units <= 100:
        bill = units * 3

    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5

    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    # classify usage based on bill amount
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    # return bill and usage status
    return bill, status


# example units consumed
units_used = 250

# calling the function
total_bill, usage_status = electricity_bill(units_used)

# display output
print("Units Consumed:", units_used)
print("Total Bill:", total_bill)
print("Usage Level:", usage_status)
```
